```python
import pandas as pd
from pathlib import Path

# 1. Detectar automáticamente la raíz del proyecto
PROJECT_ROOT = Path().resolve()
while not (PROJECT_ROOT / "src").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
    PROJECT_ROOT = PROJECT_ROOT.parent

# 2. Construir las rutas absolutas
base_dir = PROJECT_ROOT / "data_lake" / "recommender"
parquet_path = base_dir / "classified_tracks.parquet"
csv_path = base_dir / "classified_tracks.csv"

def clean_corrupted_file(file_path):
    if not file_path.exists():
        print(f"❌ No se encontró: {file_path}")
        return
    
    # 1. Leer el archivo (Parquet o CSV)
    if file_path.suffix == '.parquet':
        df = pd.read_parquet(file_path)
    else:
        df = pd.read_csv(file_path, low_memory=False)
        
    print(f"📄 Filas originales en {file_path.name}: {len(df)}")
    
    # 2. Forzar conversión a número de una columna que DEBE ser numérica.
    df['danceability'] = pd.to_numeric(df['danceability'], errors='coerce')
    
    # 3. Eliminar las filas que se hayan quedado nulas por culpa del texto cruzado
    df_clean = df.dropna(subset=['danceability'])
    
    # 4. Eliminar también las pruebas manuales recientes por si acaso
    if 'track_genre' in df_clean.columns:
        df_clean = df_clean[df_clean['track_genre'] != 'manual_input']
    
    # 5. Sobrescribir el archivo ya limpio
    if file_path.suffix == '.parquet':
        df_clean.to_parquet(file_path, index=False)
    else:
        df_clean.to_csv(file_path, index=False)
        
    print(f"✅ Filas tras limpiar {file_path.name}: {len(df_clean)} (Eliminadas: {len(df) - len(df_clean)})\n")

# Ejecutar limpieza en ambos archivos
clean_corrupted_file(parquet_path)
clean_corrupted_file(csv_path)
```
